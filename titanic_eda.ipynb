```python
# Import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Set seaborn style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (8, 5)
```

```python
# Load dataset from URL
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
```

```python
# Select required columns
df = df[['PassengerId', 'Name', 'Age', 'Gender', 'Class', 'Survived']]
print(df)
```

```python
# Display first few rows
print("First 5 Rows of Dataset:")
print(df.head())
```

```python
# 1 Countplot: Survived vs Not Survived
sns.countplot(data=df, x='Survived', palette='Set2')
plt.title("Countplot: Survived vs Not Survived")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

```python
# 2 Bar Plot: Survival by Gender and Class
sns.countplot(data=df, x='Gender', hue='Class', palette='pastel')
plt.title("Bar Plot: Gender vs Class Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title="Class")
plt.show()

```

```python
# 3 Pie Chart: Class Distribution
class_counts = df['Class'].value_counts()
labels = ['Class ' + str(i) for i in class_counts.index]
colors = ['gold', 'lightcoral', 'skyblue']

plt.pie(class_counts, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140)
plt.title("Pie Chart: Passenger Class Distribution")
plt.axis('equal')
plt.show()
```

```python
# 4 Histogram: Age Distribution
plt.hist(df['Age'].dropna(), bins=30, color='mediumorchid', edgecolor='black')
plt.title("Histogram: Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```
